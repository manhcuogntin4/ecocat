```python
# Remove columns with a high percentage of missing values
missing_threshold = 0.5
df_cleaned = df.loc[:, df.isnull().mean() < missing_threshold]
```

```python
# Remplir les valeurs manquantes restantes avec une stratégie (exemple : moyenne ou mode)
for col in df_cleaned.columns:
    if df_cleaned[col].dtype == "object":
        # Remplir les valeurs manquantes avec le mode pour les colonnes catégoriques
        df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)
    elif df_cleaned[col].dtype in ["float64", "int64"]:
        # Remplir les valeurs manquantes avec la moyenne pour les colonnes numériques
        df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
```

```python
# Identify and handle outliers in numeric columns
numeric_columns = df_cleaned.select_dtypes(include=["float64", "int64"]).columns
for col in numeric_columns:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned[col] = np.where(
        df_cleaned[col] < lower_bound, lower_bound,
        np.where(df_cleaned[col] > upper_bound, upper_bound, df_cleaned[col])
    )
```
